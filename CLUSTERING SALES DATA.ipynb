```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("fmcg_sales_data.csv")  # Adjust the path if needed

# Select numerical features for clustering
X = df[['Revenue', 'Marketing_Spend']].values

# Create and fit the KMeans model
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Get cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Plot the clustered data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', label='Data Points')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=200, linewidths=3, label='Centroids')
plt.xlabel('Revenue')
plt.ylabel('Marketing Spend')
plt.title('K-Means Clustering on FMCG Sales Data')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Print inertia (sum of squared distances to centroids)
print(f"Inertia: {kmeans.inertia_:.2f}")
```
